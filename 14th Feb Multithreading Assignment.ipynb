```python
#Multithreading Assignment
"""). What is multithreading in python? Why is it used? Name the module used to handle threads in python"""
Ans: 
Multithreading in Python is the ability to run multiple threads concurrently. Threads are lightweight 
processes that can run in parallel with other threads. They are used to improve the performance of 
applications by utilizing multiple cores of a processor. The module used to handle threads in Python 
is the threading module.

The threading module provides a simple and powerful mechanism for creating and managing threads. It 
allows multiple threads to exist within a single process, which can be used to improve the performance 
of applications that rely on concurrent execution. The threading module also provides synchronization primitives such as locks, semaphores, and condition variables to help coordinate access to shared resources between threads.

Multithreading is used to improve the performance of applications by utilizing multiple cores of a 
processor. It can also be used to make applications more responsive by allowing tasks to be performed 
in the background while the user interacts with the application.

"""Q2. Why threading module used? write the use of the following functions:"""
Ans: 
"A) activecount()-" The activeCount() method of the threading module is used to get the number of 
active threads in the current thread group. This method returns an integer value which is the number of 
threads that are currently running.

This method is useful for debugging purposes, as it can help identify any potential issues with the 
application's threading implementation. It can also be used to monitor the performance of an 
application, as it can provide insight into how many threads are actively running at any given time.

In addition, this method can be used to determine the total number of threads that have been created 
in the current thread group. This can be useful for determining the maximum number of threads that can
be created in a given application.

"currentThread()"- The currentThread() threading module is used to access the current thread object. 
It can be used to get the name of the current thread, set its name, and check if it is alive. It can 
also be used to access the main thread object, which is the first thread created when a program starts.

The currentThread() threading module is useful for debugging and logging purposes, as it can be used to
identify which thread is currently executing code. It can also be used to ensure that certain operations
are only performed on the main thread, or to ensure that certain operations are only performed on a 
specific thread.

"enumerate()"- The enumerate() function in the threading module is used to keep track of the active threads in a program. 
It returns a list of tuples containing the thread's identifier and the corresponding thread object. This can be used to monitor
and manage threads, as well as to identify which threads are running and which are not.

Enumerate() is also useful for debugging purposes, as it can help identify which threads are causing problems and which are 
not. Additionally, it can be used to ensure that all threads have completed their tasks before the program exits.

"""Q3. Explain the following functions"""
Ans: 
A) run(): The run() method is the entry point for a thread. It is the method that is called when we start a thread by 
calling the start() method. The run() method should contain the code that we want to execute in a separate thread.

B) start(): The start() method is used to start a thread by calling the run method associated with the Thread class. 
It can be called only once for any given thread object. It arranges for the object's run() method to be invoked in a separate 
thread of control.

The syntax for start() method is given below.

thread_object.start()

Example:

import threading 
  
def print_cube(num): 
    """ 
    function to print cube of given num 
    """
    print("Cube: {}".format(num * num * num)) 
  
def print_square(num): 
    """ 
    function to print square of given num 
    """
    print("Square: {}".format(num * num)) 
  
if __name__ == "__main__": 
    # creating thread 
    t1 = threading.Thread(target=print_square, args=(10,)) 
    t2 = threading.Thread(target=print_cube, args=(10,)) 
  
    # starting thread 1 
    t1.start() 
    # starting thread 2 
    t2.start() 
  
    # wait until thread 1 is completely executed 
    t1.join() 
    # wait until thread 2 is completely executed 
    t2.join() 
  
    # both threads completely executed 
    print("Done!")

Output:
Square: 100
Cube: 1000

C) Join()- The join() method in Python is used to wait for a thread to finish its execution. It blocks the calling thread 
until the thread whose join() method is called is terminated. All the threads that are created in a program, except the main 
thread, must be joined so that the program does not  terminate before all the threads complete their execution.

The join() method takes an optional argument which is a timeout value in seconds. If the timeout value is specified, the 
calling thread will wait for the thread to finish its execution or until the timeout occurs, whichever comes first. If the 
timeout value is not specified, then the calling thread will wait  indefinitely for the thread to finish its execution.

d) The threading.Thread.isAlive() method is used to check whether a thread is still executing or not. It returns True if the thread is alive and False if it is dead. This method can be used to check the status of a thread before performing any operation on it.

Syntax: threading.Thread.isAlive()

Example:

import threading 
  
def print_square(num): 
    print(num * num) 
  
 # creating thread 
t1 = threading.Thread(target=print_square, args=(10, )) 
  
# starting thread 1 
t1.start() 
  
# checking if thread is alive 
if t1.isAlive(): 
    print("Thread 1 is alive") 
  
 #wait until thread 1 is completely executed
t1.join() 
  
# both threads completely executed 
print("Done!")

"""Q4.Write a python program to create two threads. Thread one must print the list of squares and thread
two must print the list of cubes"""
Ans: 
    
    import threading
    def print_square(numbers): 
    for num in numbers: 
        print(num * num) 
  
def print_cube(numbers): 
    for num in numbers: 
        print(num * num * num) 
  
 if __name__ == "__main__": 
    arr = [2, 3, 8, 9] 
  
    t1 = threading.Thread(target=print_square, args=(arr,)) 
    t2 = threading.Thread(target=print_cube, args=(arr,)) 
  
    t1.start() 
    t2.start() 
  
    t1.join() 
    t2.join() 
  
# Output 
# 4
# 9
# 64
# 81
# 8
# 27
# 512
# 729

"""Q5.State advantages and disadvantages of multithreading"""
Ans: 
Advantages: 
1. Multithreading allows multiple tasks to be executed concurrently, which can improve overall system throughput and performance.
2. It also allows a single process to have multiple threads of execution, which can help with multitasking and parallel processing.
3. Multithreading can also help with resource sharing, as multiple threads can access the same resources.
4. It can also help with responsiveness, as multiple threads can be used to handle user input while other threads are performing 
background tasks.

Disadvantages: 
1. Multithreading can be difficult to debug and maintain, as it can be hard to keep track of multiple threads of execution.
2. It can also lead to race conditions and deadlocks, which can cause the system to become unresponsive or crash.
3. Multithreading can also lead to increased memory usage, as each thread requires its own stack and other resources.
4. It can also lead to decreased performance, as context switching between threads can be expensive.

"""Q6. Explain deadlocks and race conditions."""
Ans: Deadlocks: A deadlock is a situation in which two or more processes are blocked forever, waiting for each other to release
a resource. In Python, this can occur when two threads try to acquire the same lock at the same time.

Race Conditions: A race condition is a situation in which the outcome of a program depends on the timing or order of events. 
In Python, this can occur when two threads try to access and modify the same data at the same time. If the threads are not 
synchronized, the data can be corrupted or lost.
```
